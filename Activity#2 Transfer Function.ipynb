A transfer function has the form:

G(s) = N(s)/D(s)

where:
N(s) = numerator polynomial
D(s) = denominator polynomial

```python
import control as ctrl
import matplotlib.pyplot as plt

G = ctrl.tf([10, 20], [1, 15, 50, 0])
```

```python
print(G)
```

```python
t, y = ctrl.step_response(G)
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid()
plt.show()
```
