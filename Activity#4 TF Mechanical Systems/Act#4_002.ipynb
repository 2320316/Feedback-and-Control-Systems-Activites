```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1]
den = [1, 2, 1]
G = ctrl.tf(num, den)
print("Transfer Function G(s):", G)
```

```python
poles = ctrl.poles(G)
print("\nPoles of G(s):", poles)
```

```python
zero = ctrl.zeros(G)
print("\nZero of G(s):", zero)
```

```python
ctrl.pzmap(G, grid=True)
plt.show()
```
