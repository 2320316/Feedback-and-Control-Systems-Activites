```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1, 1, 2]
den = [1, 2, 3, 1, 1]
G = ctrl.tf(num, den)
print("Transfer Function G(s):", G)
```

```python
poles = ctrl.poles(G)
print("Poles of G(s):", poles)
```

```python
zereos = ctrl.zeros(G)
print("Zeros of G(s):", zereos)
```

```python
t, y = ctrl.step_response(G)
plt.plot(t, y) #type: ignore
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid()
plt.show()
```

```python
ctrl.pzmap(G, grid=True)
plt.title("Pole-Zero Map of G(s)")
plt.show()
```
