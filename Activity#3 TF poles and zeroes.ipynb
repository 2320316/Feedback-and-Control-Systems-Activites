```python
import control as ctrl
import matplotlib.pyplot as plt

num = [10, 20]
den = [1, 15, 50, 0]

G = ctrl.tf(num, den)
print(G)
```

Poles are the roots of the denominator

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

zeros are the roots of the numerator

```python
zeros = ctrl.zeros(G)
print("Zeros:", zeros)
```

```python
ctrl.pzmap(G, grid=True)
plt.show()
```
